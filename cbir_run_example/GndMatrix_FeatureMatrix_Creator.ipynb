```python
#from list P take each image name change it to an index from 0 to 6023 e.g 'paris_defense_000004\n' to 1

def NamesToIndex(X):
    A=[]
    co2=0

    for i in X:
     

        
        count=0

        for j in P:

            if (i==j):

                A.append(count)
                continue

            count+=1

            
    A = np.sort(A)
    return A
```

```python
#creat gand truth matrix from grand truth files given with the dataset which contain each query image corresponding images that should be retrived 
idx={}
X=[]
gnd_t=[]
listA=["defense" ,"eiffel","invalides","louvre",'moulinrouge','museedorsay','notredame','pantheon','pompidou','sacrecoeur','triomphe' ]
listB=["good","junk","ok"]
for i in listA:
    for j in range(1,6):
        idx={}
        for z in listB: 
            
            path = ('D:/parisgt/{}_{}_{}.txt').format(i,j,z)
            print('D:/parisgt/{}_{}_{}.txt'.format(i,j,z))#"ok"" junk" 
            file = open(path,'r')
            X=file.readlines()
            file.close()
            idx[z]=(NamesToIndex(X))
        gnd_t.append(idx)
        
    print('#')
```

```python
#save grand truth table matrix as a pickle file
pickle_out=open('D:/ssd/1/effcientnet1A/gnd_t.pickle','wb')
pickle.dump(gnd_t,pickle_out)
pickle_out.close()
```
